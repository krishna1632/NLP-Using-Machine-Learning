# Stemming


```python
words = [
    "eating",
    "eats",
    "eaten",
    "writing",
    "writes",
    "programming",
    "programs",
    "history",
    "finally",
    "finalized",
]
```

### `PorterStemmer` Class


```python
from nltk.stem import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
for word in words:
    print(f"{word} ----> {porter_stemmer.stem(word)}")
```

```python
porter_stemmer.stem("Marriage")
```

```python
porter_stemmer.stem('fairly')
```

### `RegexpStemmer` Class


```python
from nltk.stem import RegexpStemmer
```

```python
regexp_stemmer = RegexpStemmer("ing$|s$|e$|able$")
```

```python
regexp_stemmer.stem("Hareing")
```

### `SnowballStemmer` Class


```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer(language="english")
```

```python
for word in words:
    print(f"{word} ----> {snowball_stemmer.stem(word)}")
```

```python
snowball_stemmer.stem('fairly')
```
