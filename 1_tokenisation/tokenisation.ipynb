# Tokenisation


```python
corpus = """Hello, I'm am Krishna Singha.
I like AI, ML and NLP.
"""

print(corpus)
```

#### 1. Corpus(Paragraph) -> Documents(Sentences)

```python
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
for sentence in documents:
    print(sentence)
```

#### 2. Corpus(Paragraph) -> Words

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(corpus)
```
